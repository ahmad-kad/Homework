```python
def coinChange(coins, amount):
    if amount < 0:
        return -1

    dp = [float('inf')] * (amount + 1)
    dp[0] = 0

    for coin in coins:
        for i in range(coin, amount + 1):
            dp[i] = min(dp[i], dp[i - coin] + 1)
    return dp[amount] if dp[amount] != float('inf') else -1
```

```python
print(coinChange([1, 2, 5], 11))  # Output: 3
print(coinChange([2], 3))          # Output: -1
print(coinChange([1], 0))  
```

```python
print(coinChange([], 10))                      # Output: -1
print(coinChange([1, 2, 5], 10000))             # Output: 2000 (assuming optimal solution)
print(coinChange([100, 200, 500], 1000))        # Output: 2 (assuming optimal solution)
print(coinChange([1, 2, 5], 1))                 # Output: 1
print(coinChange([1, 2, 5], 0))                 # Output: 0
print(coinChange([10], 5))                      # Output: -1
print(coinChange([1, 1, 1, 1, 1], 5))           # Output: 5
print(coinChange([1, 2, 5], -10))               # Output: -1
print(coinChange([1] * 12, 10))                 # Output: 10
```
